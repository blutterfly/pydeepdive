```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Convert 'Portfolio%' from string to float
df1['Portfolio%'] = df1['Portfolio%'].str.rstrip('%').astype('float') / 100.0

# Top 10 Companies by Portfolio Percentage
top_10 = df1.nlargest(10, 'Portfolio%')

# Pie Chart: Portfolio Percentage
plt.figure(figsize=(10, 7))
plt.pie(top_10['Portfolio%'], labels=top_10['Company'], autopct='%1.1f%%')
plt.title('Portfolio Percentage of Top 10 Companies')
plt.show()
```

```python
# Bar Chart: Price Change
plt.figure(figsize=(12, 8))
sns.barplot(x='Chg', y='Company', data=top_10)
plt.title('Price Change of Top 10 Companies')
plt.xlabel('Change in Price')
plt.ylabel('Company')
plt.show()
```

```python
# Horizontal Bar Chart: Company Prices
plt.figure(figsize=(12, 8))
sns.barplot(x='Price', y='Company', data=top_10, orient='h')
plt.title('Company Prices of Top 10 Companies')
plt.xlabel('Price')
plt.ylabel('Company')
plt.show()
```
